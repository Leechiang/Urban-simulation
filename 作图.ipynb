```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 读取CSV文件
data = pd.read_csv("C:/Users/86180/Desktop/US/cw/作图用.csv")
```

```python
# 删除关于diameter的列
data = data.drop(columns=data.filter(like='diameter').columns)

# 提取需要的列
s1_efficiency_columns = data.filter(regex='s1_(closeness|degree|betweenness)_efficiency')
s1_largestC_columns = data.filter(regex='s1_(closeness|degree|betweenness)_largestC')
s2_efficiency_columns = data.filter(regex='s2_(closeness|degree|betweenness)_efficiency')
s2_largestC_columns = data.filter(regex='s2_(closeness|degree|betweenness)_largestC')

# 绘制折线图
plt.figure(figsize=(12, 6))

# 绘制s1的efficiency图
ax1 = plt.subplot(1, 2, 1)
for column in s1_efficiency_columns:
    plt.plot(data.index, data[column], label=column)
plt.title('s1 Efficiency')
plt.xlabel('Removed Nodes')
plt.ylabel('Efficiency')
plt.xticks(data.index, range(1, 11))

# 添加左侧纵坐标轴
ax2 = ax1.twinx()
for column in s1_largestC_columns:
    plt.plot(data.index, data[column], linestyle='--', label=column)

# 绘制s2的efficiency图
ax3 = plt.subplot(1, 2, 2)
for column in s2_efficiency_columns:
    plt.plot(data.index, data[column], label=column)
plt.title('s2 Efficiency')
plt.xlabel('Removed Nodes')
plt.ylabel('Efficiency')
plt.xticks(data.index, range(1, 11))

# 添加右侧纵坐标轴
ax4 = ax3.twinx()
for column in s2_largestC_columns:
    plt.plot(data.index, data[column], linestyle='--', label=column)

# 绘制图例
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax3.get_legend_handles_labels()
handles = handles1 + handles2
labels = labels1 + labels2
plt.legend(handles, labels, loc='lower center', bbox_to_anchor=(0.5, -0.3), fancybox=True, shadow=True, ncol=2)

plt.tight_layout()
plt.show()
```
